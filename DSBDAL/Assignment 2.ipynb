```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("student_acedamic.csv")
```

```python
df.isna().sum()
```

```python
df.describe()
```

```python
ndf=df
```

```python
ndf.dropna(subset=["Math Score"],inplace=True)
```

```python
ndf.isna().sum()
```

```python
mean_w=ndf["Writing Score"].mean()
```

```python
ndf["Writing Score"].fillna(mean_w,inplace=True)
```

```python
import numpy as np
ndf=df
ndf["Reading Score"].replace(np.nan,75,inplace=True)
```

```python
ndf.sample(10)
```

```python

plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Placement Score'])
sns.swarmplot(x=df['Placement Score'], color='red', marker='o', size=5)
plt.xlabel('Placement Score')
plt.title('Box Plot with Outliers')
plt.show()
```

```python
lower_threshold = 75
upper_threshold = 100
outliers = (ndf['Placement Score'] < lower_threshold) | (ndf['Placement Score'] > upper_threshold)
```

```python
mean_value = ndf['Placement Score'].mean()
```

```python
ndf.loc[outliers, 'Placement Score'] = mean_value
```
